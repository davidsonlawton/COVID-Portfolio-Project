```python
df = df.set_index('id')
```

```python
df = df[['title', 'description', 'publishTime', 
         'duration_seconds', 'viewCount', 'likeCount',
        'commentCount']]
```

```python
df.head()
```

```python
df.shape
```

```python
df_subset1 = df.query('viewCount > 1_000_000')
```

```python
df = df.loc[~df['likeCount'].isna()]
```

```python
df.dtypes
```

```python
df['viewCount'] = df['viewCount'].astype('int')
df['likeCount'] = df['likeCount'].astype('int')
```

```python
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
df['likeCount'] = pd.to_numeric(df['likeCount'].astype('str'))
```

```python
df['like_to_view_ratio'] = df['likeCount'] / df['viewCount']
```

```python
df['viewCount'].plot(kind = 'hist', color = 'red', bins = 50, 
                     title = 'Distribution of View Count', figsize = (15, 5))
```

```python
df.plot(kind = 'scatter', x = 'viewCount',
       y = 'likeCount', title = 'View vs Like Count', color = 'red')
```

```python
df.to_csv('processed_MrBeastdata.csv')
```
